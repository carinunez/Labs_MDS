La columna `Is An Alien` deber칤a ser de tipo boolean, `Transaction ID` deber칤a ser de tipo double y `Transaction Hour` deber칤a ser de tipo timestamp_ntz.

````python
from pyspark.sql.types import DoubleType, BooleanType, DayTimeIntervalType

def cast_columns(df, cols_types):
    """
    Cambia el tipo de m칰ltiples columnas en un DataFrame de Spark.

    Par치metros:
    - df: DataFrame de Spark.
    - cols_types: Diccionario con nombres de columnas como claves y tipos de
                    datos de Spark como valores.

    Retorna:
    - DataFrame de Spark con tipos de columnas modificados.
    """
    for column, col_type in cols_types.items():
      df = df.withColumn(column, df[column].cast(col_type))

    return df
````

````python
df = cast_columns(df, {'Is An Alien': BooleanType()})

df.dtypes
````

````python
print('---------  P6 Contar nulos por variable ----------')
columns = df.columns

for col in columns:
  conteo_nulos = df.na.drop(how="any",subset=[col]).count()
  print(col + ' : ', conteo_nulos)
df.show()

print('---------  P7 Borrar datos nulos  ----------')

for col in columns:
  df = df.na.drop(how="any",subset=[col])
df.show()

print('---------  P8 Borrar duplicados  ----------')
df = df.dropDuplicates()
df.show()
````

3.   Crear la variable *Type of purchase* seg칰n la catidad de unidades vendidas de acuerdo a las siguientes categor칤as.
  * Compra minorista: 4 productos o menos.
  * Compra mayorista: 5 produtos o m치s.

````python
import pyspark.sql.functions as F

# print('---------  P1 Bodoque pesos log(48 +|x|^36)  ----------')
df = df.withColumn('Transaction_bp', F.log(48 + F.abs((df['Transaction Amount'])**36)))

# print('---------  P2 Transaction Month  ----------')
df = df.withColumn('Transaction Month', F.month('Transaction Date'))

# print('---------  P3 Type of purchase (minorista <=4)  ----------')

# df.filter('Quantity<=4').show()
# df.filter('Quantity>=5').show()

df = df.withColumn('Type of purchase',
                    F.when(df['Quantity']<=4, 'Compra minorista')\
                    .when(df['Quantity']>=5, 'Compra mayorista'))

print('---------  P4 compras alienigenas en comercio mayorista  ----------')
df.filter(df['Is An Alien']==1.0)\
  .filter(df['Type of purchase']=='Compra mayorista')
df.show()

print('---------  P5  cant compras humanas (false) y alienigenas (true)  ----------')
df.groupBy('Is An Alien').count().show()

print('---------  P6 Tabla de recaudaci칩n prom por metodo de pago ----------')

df_plot = df.groupBy('Payment Method').pivot('Is An Alien').mean("Transaction Amount")
df_plot.show()
````

Adem치s, est치n interesados en aprovechar las funciones de ventana en SQL para an치lisis avanzados. Las funciones de ventana permiten realizar c치lculos sobre un conjunto de filas que est치n relacionadas con la fila actual. Por ejemplo, UNBOUNDED PRECEDING se usa para indicar que el rango de la funci칩n de ventana comienza desde la primera fila de la partici칩n o del conjunto de resultados, lo cual es 칰til para calcular sumas acumulativas hasta la fila actual. Las variaciones comunes de este uso incluyen:

- `UNBOUNDED PRECEDING` to `CURRENT ROW`: Calcula desde el inicio de la partici칩n hasta la fila actual.
- `UNBOUNDED PRECEDING` to `UNBOUNDED FOLLOWING`: Cubre todas las filas dentro de la partici칩n.
- `VALUE PRECEDING` to `VALUE FOLLOWING`: Establece un rango espec칤fico basado en valores antes y despu칠s de la fila actual.
