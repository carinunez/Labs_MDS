````python
def exploratory_data_analysis(dataframe):
    # Imprimir dim DataFrame
    print('Dimensiones del dataframe: ',dataframe.shape)
    print('------------------------------------------------------')
    # Imprimir nombre cols
    print('Columnas: ', list(dataframe.columns))
    print('------------------------------------------------------')

    # Mostrar head(5) tail(5) y 5 datos random
    print('Primeros 5 elementos')
    display(dataframe.head(5))

    print('Ultimos 5 elementos')
    display(dataframe.tail(5))
    print('5 elementos aleatorios')
    display(dataframe.sample(5))

    print('------------------------------------------------------')
    # mostrar descripci√≥n variables .describe()
    
    # Dado que las columnas que tienen variables num√©ricas de inter√©s son de tipo 
    # objeto, antes de cambiar su formato, es necesario buscar las columnas que contengan 
    # la palabra 'score' en su nombre. 
    print('Descripci√≥n de las variables num√©ricas')
    cols_score =  dataframe.filter(like='score').columns
    dataframe[cols_score] = dataframe[cols_score].apply(pd.to_numeric)
    
    display(dataframe.describe())

    print('------------------------------------------------------')
    # imprimir num valores nulos -> isna().sum()
    print('Valores nulos')
    print(dataframe.isna().sum())

    print('------------------------------------------------------')
    # imprimir num valores unicos -> -nunique()
    print('Valores unicos')
    print(dataframe.nunique())

    print('------------------------------------------------------')
    # imprimir total filas con valores duplicados .duplicated(subset= ALGO)
    print('Total de filas duplicadas')
    print(dataframe.duplicated(subset='names').astype(int).sum())

    print('------------------------------------------------------')
    print('Nombres duplicados')
    display(dataframe.loc[dataframe.duplicated(subset='names', keep='first')].sort_values('names'))
    pass
````

````python
# Escriba su c√≥digo aqu√≠
def clean_data(dataframe):
    # Eliminar registros con valores nulos.
    dataframe.dropna(inplace=True)

    # Eliminar valores fuera del rango v√°lido.
    cols_score =  dataframe.filter(like='score').columns
    
    for col in cols_score:
        index_del = dataframe.loc[(dataframe[col]>100)|(dataframe[col]<0)].index
        dataframe.drop(index=index_del, axis=0, inplace=True)

    # Eliminar registros con nombres duplicados
    dataframe.drop_duplicates(subset='names', keep='first', inplace=True)
    
    return dataframe
````

### 3.1 Transformaci√≥n de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intenci√≥n de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci√≥n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci√≥n entregada, cree una funci√≥n que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
# Escriba su c√≥digo aqu√≠
def nota_chilena(dataframe_in, columna):
  dataframe_in[columna] = dataframe_in[columna].apply(lambda x: np.round(x*0.06 + 1.0, 2))
  return dataframe_in
````

Contentos con su desempe√±o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a√±o y los que han tenido desempe√±o de destacado. Cree una funci√≥n que permita obtener tres Dataframes:
- El primero deber√° poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser√° un Data Frame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
#Escriba su c√≥digo aqu√≠
def reprobados_destacados(dataframe_in):
    # retorna 3 df
    reprobados = dataframe_in.loc[dataframe_in['Promedio']< 4.0]

    reprobados_r = reprobados.copy()
    reprobados_r['Promedio'] = 'R'

    # notas > 5.8 en todas las asignaturas
    destacados = dataframe_in.loc[(dataframe_in['math score']>5.8)|
                                  (dataframe_in['reading score']>5.8)|
                                  (dataframe_in['writing score']>5.8)]
    destacados = destacados.sort_values('Promedio', ascending=False)
    return reprobados, reprobados_r, destacados
````
